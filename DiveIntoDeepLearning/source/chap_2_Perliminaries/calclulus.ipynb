# 1. 绘制函数$y = f(x) = x^3 - \frac{1}{x}$和其在$x = 1$处切线的图像。

```python
f = lambda x: x**3 - 1/x
f_prime = lambda x: 3*x**2 + 1/x**2
def f_tangent(X, x):
    """
    X = 当前x点的位置
    x = 求解斜率的x点
    """
    slope = f_prime(x)
    y = f(x)
    return slope*(X - x) + y

print(f_tangent(3, 1))

x = np.arange(0.1, 3.0, 0.1)
d2l.plot(x, Y=(f(x), f_tangent(x, 1)))
```

# 2. 求函数$f(\mathbf{x}) = 3x_1^2 + 5e^{x_2}$的梯度。

```python
f = lambda x_1, x_2: 3*x_1 + 5*np.exp(1)**x_2
f_x_1_partial = lambda x: 6*x
f_x_2_partial = lambda x: 5*np.exp(1)**x
def get_gradient(f, x):
    x_1_partial = f_x_1_partial(x[0])
    x_2_partial = f_x_2_partial(x[1])
    return x_1_partial, x_2_partial
get_gradient(f, [1, 2])
```
